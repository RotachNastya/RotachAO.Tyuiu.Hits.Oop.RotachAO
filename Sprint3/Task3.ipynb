```
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Color { get; set; }          // Цвет животного
    public double Weight { get; set; }         // Вес животного в кг
    public string Diet { get; set; }          // Тип питания (хищник, травоядный, всеядный)

    public Animal(string name, int age, string color, double weight, string diet)
    {
        Name = name;
        Age = age;
        Color = color;
        Weight = weight;
        Diet = diet;
    }

    // Виртуальный метод для звука
    public virtual string Sound() => "Издает звук";

    // Метод питания
    public virtual string Eat() => $"{Name} Ест.";

    // Метод питания с указанием конкретной еды
    public virtual string Eat(string food) => $"{Name} Ест {food}.";

    // Метод сна
    public virtual string Sleep() => $"{Name} Спит.";

    // Метод перемещения
    public virtual string Move() => $"{Name} Двигается.";

    public override string ToString()
    {
        return $"Имя: {Name}, Возраст: {Age}, Цвет: {Color}, Вес: {Weight}кг, Рацион: {Diet}";
    }
}

// Простое наследование - собака наследуется напрямую от животного
public class Dog : Animal
{
    public string Breed { get; set; }          // Порода собаки (например, "овчарка", "лабрадор")
    public bool IsTrained { get; set; }          // Признак дрессированности собаки

    public Dog(string name, int age, string color, double weight, string diet, string breed, bool isTrained)
        : base(name, age, color, weight, diet)
    {
        Breed = breed;
        IsTrained = isTrained;
    }

    // Переопределение метода Sound
    public override string Sound() => "Гав!";

    // Переопределение метода Move
    public override string Move() => $"{Name}  бегает.";

    public string Guard() => $"{Name}  охраняет.";

    public override string ToString()
    {
        return base.ToString() + $", Порода: {Breed}, тренируется: {IsTrained}";
    }
}

// Сложное наследование через промежуточный класс
public abstract class Mammal : Animal
{
    public bool HasFur { get; set; }           // Наличие шерсти (true - есть шерсть, false - нет шерсти)
    public int GestationPeriod { get; set; }   // Период беременности в днях

    public Mammal(string name, int age, string color, double weight, string diet, bool hasFur, int gestationPeriod)
        : base(name, age, color, weight, diet)
    {
        HasFur = hasFur;
        GestationPeriod = gestationPeriod;
    }

    public string FeedMilk() => "Кормит молоко.";
}

public class Cat : Mammal
{
    public Cat(string name, int age, string color, double weight, string diet, bool hasFur, int gestationPeriod)
        : base(name, age, color, weight, diet, hasFur, gestationPeriod) { }

    // Переопределение метода Sound
    public override string Sound() => "Мяу!";

    // Переопределение метода Move
    public override string Move() => $"{Name} is climbing.";

    public string Climb() => $"{Name}  залазит";

    public override string ToString()
    {
        return base.ToString() + $", Имеет мех: {HasFur}, GestationPeriod: {GestationPeriod}";
    }
}

// Множественное наследование через интерфейсы
public interface IPet
{
    string Play();  // Интерфейс для домашних питомцев
}

public interface ITrainable
{
    string Train(); // Интерфейс для животных, которых можно дрессировать
}

public class SmartDog : Dog, IPet, ITrainable
{
    public SmartDog(string name, int age, string color, double weight, string diet, string breed, bool isTrained)
        : base(name, age, color, weight, diet, breed, isTrained) { }

    public string Play() => $"{Name}  играет.";  // Реализация интерфейса IPet
    public string Train() => $"{Name} тренируется"; // Реализация интерфейса ITrainable
}

// Generic класс для создания вольеров для животных с определенным типом питания
public class Enclosure<T> where T : Animal
{
    private List<T> animals = new List<T>();  // Список животных в вольере
    public string EnclosureType { get; set; } //  Укажем тип вальера

    public Enclosure(string enclosureType)
    {
        EnclosureType = enclosureType;
    }

    // Добавление животного в вольер
    public void AddAnimal(T animal)
    {
        animals.Add(animal);
        Console.WriteLine($"{animal.Name} помещен в {EnclosureType}");
    }

    // Заставить всех животных в вольере издать звук
    public void MakeAllSounds()
    {
        Console.WriteLine($"Звуки в {EnclosureType}:");
        foreach (var animal in animals)
        {
            Console.WriteLine($"{animal.Name} говорит: {animal.Sound()}");
        }
    }

     public override string ToString()
    {
        string animalList = "";
        foreach(var animal in animals) {
            animalList += "\t" + animal.ToString() + "\n";
        }

        return $"Enclosure Type: {EnclosureType}\nAnimals:\n{animalList}";

    }

}


        // Создаем объекты
        var dog = new Dog("Рекс", 3, "Коричневый", 15.5, "Mясо", "Такса", true);
        var cat = new Cat("Барсик", 2, "Белый", 4.2, "Mясо", true, 63);
        var smartDog = new SmartDog("Maкс", 2, "Черный", 12.0, "Mясо", "Лабрадор", true);

        // Демонстрация полиморфизма
        List<Animal> animals = new List<Animal>()
        {
            dog,
            cat,
            smartDog
        };

        Console.WriteLine("Demonstration of Polymorphism:");
        foreach (var animal in animals)
        {
            Console.WriteLine(animal.GetType().Name + " " + animal.Name + ":");
            Console.WriteLine($"\tИздает звук: {animal.Sound()}");
            Console.WriteLine($"\tЕст: {animal.Eat()}");
            Console.WriteLine($"\tСпит: {animal.Sleep()}");
            Console.WriteLine($"\tДвижется: {animal.Move()}");
            Console.WriteLine($"\tИнформация: {animal}");

            if(animal is Dog currentDog)
            {
                Console.WriteLine($"\tОхраняет: {currentDog.Guard()}");
            }

            Console.WriteLine("------------------");
        }

        // Демонстрация Generic класса
        var predatorEnclosure = new Enclosure<Animal>("Вольер для хищников");  // Все животные
        predatorEnclosure.AddAnimal(dog);
        predatorEnclosure.AddAnimal(cat);   // Все животные хищники
        Console.WriteLine("\n" + predatorEnclosure.ToString());
        predatorEnclosure.MakeAllSounds();

```
