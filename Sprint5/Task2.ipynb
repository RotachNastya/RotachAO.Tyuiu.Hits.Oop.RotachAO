````
using System;
using System.Collections.Generic;

// –î–µ–ª–µ–≥–∞—Ç –¥–ª—è –¥–µ–π—Å—Ç–≤–∏–π –∂–∏–≤–æ—Ç–Ω—ã—Ö
public delegate void AnimalAction(string message);

// –î–µ–ª–µ–≥–∞—Ç –¥–ª—è –∑–≤—É–∫–æ–≤ –∂–∏–≤–æ—Ç–Ω—ã—Ö
public delegate void MakeSoundDelegate();

public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }
    
    // –°–æ–±—ã—Ç–∏–µ –¥–ª—è —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π
    public event AnimalAction OnAction;
    
    // –î–µ–ª–µ–≥–∞—Ç –¥–ª—è –∏–∑–¥–∞–Ω–∏—è –∑–≤—É–∫–∞
    public MakeSoundDelegate MakeSound;

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"–ò–º—è: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age} –ª–µ—Ç");
    }

    // –ú–µ—Ç–æ–¥ –¥–ª—è –≤—ã–∑–æ–≤–∞ —Å–æ–±—ã—Ç–∏—è
    protected void RaiseAction(string message)
    {
        OnAction?.Invoke(message);
    }

    public void PerformAction(string action)
    {
        RaiseAction($"{Name} –≤—ã–ø–æ–ª–Ω—è–µ—Ç –¥–µ–π—Å—Ç–≤–∏–µ: {action}");
    }
}

public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed)
        : base(name, age)
    {
        Breed = breed;
        
        // –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ–ª–µ–≥–∞—Ç–∞ –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º
        MakeSound = () => {
            Console.WriteLine($"–°–æ–±–∞–∫–∞ {Name} –ª–∞–µ—Ç: –ì–∞–≤-–≥–∞–≤!");
            RaiseAction($"{Name} –ª–∞–µ—Ç");
        };
        
        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏–µ —á–µ—Ä–µ–∑ –ª—è–º–±–¥—É
        OnAction += (message) => Console.WriteLine($"üêï –£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ: {message}");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–ü–æ—Ä–æ–¥–∞: {Breed}");
    }
}

public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor)
        : base(name, age)
    {
        IsIndoor = isIndoor;
        
        // –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ–ª–µ–≥–∞—Ç–∞ –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º
        MakeSound = () => {
            Console.WriteLine($"–ö–æ—à–∫–∞ {Name} –º—è—É–∫–∞–µ—Ç: –ú—è—É-–º—è—É!");
            RaiseAction($"{Name} –º—è—É–∫–∞–µ—Ç");
        };
        
        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏–µ —á–µ—Ä–µ–∑ –ª—è–º–±–¥—É
        OnAction += (message) => Console.WriteLine($"üêà –£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ: {message}");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–î–æ–º–∞—à–Ω—è—è –∫–æ—à–∫–∞: {IsIndoor}");
    }
}

public class Bird : Animal
{
    public double Wingspan { get; set; }

    public Bird(string name, int age, double wingspan)
        : base(name, age)
    {
        Wingspan = wingspan;
        
        // –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ–ª–µ–≥–∞—Ç–∞ –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º
        MakeSound = () => {
            Console.WriteLine($"–ü—Ç–∏—Ü–∞ {Name} –ø–æ–µ—Ç: üéµ –ß–∏–∫-—á–∏—Ä–∏–∫!");
            RaiseAction($"{Name} –ø–æ–µ—Ç");
        };
        
        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏–µ —á–µ—Ä–µ–∑ –ª—è–º–±–¥—É
        OnAction += (message) => Console.WriteLine($"üê¶ –£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ: {message}");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–†–∞–∑–º–∞—Ö –∫—Ä—ã–ª—å–µ–≤: {Wingspan} —Å–º");
    }
}

public class Fish : Animal
{
    public string WaterType { get; set; }

    public Fish(string name, int age, string waterType)
        : base(name, age)
    {
        WaterType = waterType;
        
        // –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ–ª–µ–≥–∞—Ç–∞ –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º
        MakeSound = () => {
            Console.WriteLine($"–†—ã–±–∫–∞ {Name} –ø—É—Å–∫–∞–µ—Ç –ø—É–∑—ã—Ä–∏: üåä –ë—É–ª—å-–±—É–ª—å!");
            RaiseAction($"{Name} –ø—É—Å–∫–∞–µ—Ç –ø—É–∑—ã—Ä–∏");
        };
        
        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏–µ —á–µ—Ä–µ–∑ –ª—è–º–±–¥—É
        OnAction += (message) => Console.WriteLine($"üê† –£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ: {message}");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–¢–∏–ø –≤–æ–¥—ã: {WaterType}");
    }
}

class Program
{
    // –°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–π –¥–µ–ª–µ–≥–∞—Ç –¥–ª—è –æ–±—â–∏—Ö —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–π
    static AnimalAction globalNotifier = (message) => 
        Console.WriteLine($"üîî –ì–ª–æ–±–∞–ª—å–Ω–æ–µ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ: {message}");

    static void Main()
    {
        Console.WriteLine("=== –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –î–ï–õ–ï–ì–ê–¢–û–í, –õ–Ø–ú–ë–î–ê-–í–´–†–ê–ñ–ï–ù–ò–ô –ò –°–û–ë–´–¢–ò–ô ===\n");

        // –°–æ–∑–¥–∞–µ–º –∂–∏–≤–æ—Ç–Ω—ã—Ö
        Dog dog = new Dog("–ë–æ–±–∏–∫", 3, "–û–≤—á–∞—Ä–∫–∞");
        Cat cat = new Cat("–ú—É—Ä–∫–∞", 2, true);
        Bird bird = new Bird("–ö–µ—à–∞", 1, 15.5);
        Fish fish = new Fish("–ù–µ–º–æ", 1, "–ú–æ—Ä—Å–∫–∞—è");

        // –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º—Å—è –Ω–∞ –≥–ª–æ–±–∞–ª—å–Ω—ã–µ —É–≤–µ–¥–æ–º–ª–µ–Ω–∏—è
        dog.OnAction += globalNotifier;
        cat.OnAction += globalNotifier;
        bird.OnAction += globalNotifier;
        fish.OnAction += globalNotifier;

        // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –ø–æ–¥–ø–∏—Å–∫–∞ —á–µ—Ä–µ–∑ –ª—è–º–±–¥—É
        dog.OnAction += (msg) => Console.WriteLine($"üíñ –õ—é–±–∏–º–µ—Ü: {msg}");

        // –°–ø–∏—Å–æ–∫ –∂–∏–≤–æ—Ç–Ω—ã—Ö
        List<Animal> animals = new List<Animal> { dog, cat, bird, fish };

        // 1. –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –¥–µ–ª–µ–≥–∞—Ç–æ–≤ - –∏–∑–¥–∞–Ω–∏–µ –∑–≤—É–∫–æ–≤
        Console.WriteLine("=== –ò–ó–î–ê–ù–ò–ï –ó–í–£–ö–û–í (–î–ï–õ–ï–ì–ê–¢–´) ===");
        foreach (var animal in animals)
        {
            animal.DisplayInfo();
            animal.MakeSound(); // –í—ã–∑–æ–≤ —á–µ—Ä–µ–∑ –¥–µ–ª–µ–≥–∞—Ç
            Console.WriteLine();
        }

        // 2. –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Å–æ–±—ã—Ç–∏–π - –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –¥–µ–π—Å—Ç–≤–∏–π
        Console.WriteLine("=== –í–´–ü–û–õ–ù–ï–ù–ò–ï –î–ï–ô–°–¢–í–ò–ô (–°–û–ë–´–¢–ò–Ø) ===");
        dog.PerformAction("–±–µ–∂–∏—Ç –∑–∞ –ø–∞–ª–∫–æ–π");
        cat.PerformAction("—Ç–æ—á–∏—Ç –∫–æ–≥—Ç–∏");
        bird.PerformAction("–ª–µ—Ç–∞–µ—Ç –ø–æ –∫–æ–º–Ω–∞—Ç–µ");
        fish.PerformAction("–ø–ª—ã–≤–µ—Ç –≤ –∞–∫–≤–∞—Ä–∏—É–º–µ");

        Console.WriteLine("\n=== –õ–Ø–ú–ë–î–ê-–í–´–†–ê–ñ–ï–ù–ò–Ø –î–õ–Ø –§–ò–õ–¨–¢–†–ê–¶–ò–ò ===");

        // –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–π –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏
        var youngAnimals = animals.FindAll(a => a.Age < 3);
        Console.WriteLine("–ú–æ–ª–æ–¥—ã–µ –∂–∏–≤–æ—Ç–Ω—ã–µ (–¥–æ 3 –ª–µ—Ç):");
        youngAnimals.ForEach(a => a.DisplayInfo());

        Console.WriteLine("\n=== –ì–†–£–ü–ü–û–í–´–ï –î–ï–ô–°–¢–í–ò–Ø –° –õ–Ø–ú–ë–î–ê–ú–ò ===");

        // –ì—Ä—É–ø–ø–æ–≤–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ —Å –ª—è–º–±–¥–æ–π
        Action<Animal> feedAnimal = (animal) => {
            Console.WriteLine($"üçñ –ö–æ—Ä–º–∏–º {animal.Name}");
            animal.PerformAction("–µ—Å—Ç");
        };

        // –ö–æ—Ä–º–∏–º –≤—Å–µ—Ö –∂–∏–≤–æ—Ç–Ω—ã—Ö
        animals.ForEach(feedAnimal);

        // –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è Func –¥–µ–ª–µ–≥–∞—Ç–∞
        Console.WriteLine("\n=== –í–´–ß–ò–°–õ–ï–ù–ò–Ø –° FUNC –î–ï–õ–ï–ì–ê–¢–û–ú ===");
        Func<Animal, string> animalDescription = (a) => 
            $"{a.Name} ({a.GetType().Name}) - {a.Age} –ª–µ—Ç";

        foreach (var animal in animals)
        {
            Console.WriteLine(animalDescription(animal));
        }

        Console.ReadLine();
    }
}
````
