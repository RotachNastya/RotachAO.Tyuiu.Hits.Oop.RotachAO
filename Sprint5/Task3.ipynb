–î–æ–±–∞–≤—å—Ç–µ –∫ —Å—É—â–µ—Å—Ç—É—é—â–∏–º –∫–ª–∞—Å—Å–∞–º (–±–∞–∑–æ–≤—ã–º—É –∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–º 3-4 –∞—Ç—Ä–∏–±—É—Ç–∞ –∏ –º–µ—Ç–æ–¥–∞) –∏—Å–æ–ª—å–∑—É–π—Ç–∏–µ –≤ –ø—Ä–æ–µ–∫—Ç–µ –∫–æ–ª–ª–µ–∫—Ü–∏–∏, –¥–µ–ª–µ–≥–∞—Ç—ã, —Å–æ–±—ã—Ç–∏—è.

```
using System;
using System.Collections.Generic;

// –î–µ–ª–µ–≥–∞—Ç –¥–ª—è –¥–µ–π—Å—Ç–≤–∏–π –∂–∏–≤–æ—Ç–Ω—ã—Ö
public delegate void AnimalAction(string message);
// –î–µ–ª–µ–≥–∞—Ç –¥–ª—è –∫–æ—Ä–º–ª–µ–Ω–∏—è
public delegate void FeedAnimalHandler(string foodType);

public class Animal
{
    // –ê—Ç—Ä–∏–±—É—Ç—ã –±–∞–∑–æ–≤–æ–≥–æ –∫–ª–∞—Å—Å–∞
    public string Name { get; set; }
    public int Age { get; set; }
    public double Weight { get; set; }
    public string Habitat { get; set; }
    public bool IsHungry { get; set; }

    // –°–æ–±—ã—Ç–∏—è
    public event AnimalAction OnAction;
    public event FeedAnimalHandler OnFeed;

    public Animal(string name, int age, double weight, string habitat)
    {
        Name = name;
        Age = age;
        Weight = weight;
        Habitat = habitat;
        IsHungry = true;
    }

    // –í–∏—Ä—Ç—É–∞–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã
    public virtual void DisplayInfo()
    {
        Console.WriteLine($"–ò–º—è: {Name}, –í–æ–∑—Ä–∞—Å—Ç: {Age}, –í–µ—Å: {Weight}–∫–≥, –°—Ä–µ–¥–∞: {Habitat}");
    }

    public virtual void Sleep()
    {
        RaiseAction($"{Name} —Å–ø–∏—Ç");
    }

    public virtual void Move()
    {
        RaiseAction($"{Name} –ø–µ—Ä–µ–º–µ—â–∞–µ—Ç—Å—è");
    }

    // –ú–µ—Ç–æ–¥ –¥–ª—è –≤—ã–∑–æ–≤–∞ —Å–æ–±—ã—Ç–∏—è
    protected void RaiseAction(string message)
    {
        OnAction?.Invoke(message);
    }

    // –ú–µ—Ç–æ–¥ –∫–æ—Ä–º–ª–µ–Ω–∏—è
    public void Feed(string foodType)
    {
        IsHungry = false;
        OnFeed?.Invoke(foodType);
        RaiseAction($"{Name} –ø–æ–∫–æ—Ä–º–ª–µ–Ω({foodType})");
    }
}

public class Dog : Animal
{
    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã
    public string Breed { get; set; }
    public bool IsTrained { get; set; }
    public string FavoriteToy { get; set; }

    public Dog(string name, int age, double weight, string breed, bool isTrained, string favoriteToy) 
        : base(name, age, weight, "–î–æ–º")
    {
        Breed = breed;
        IsTrained = isTrained;
        FavoriteToy = favoriteToy;

        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏—è —á–µ—Ä–µ–∑ –ª—è–º–±–¥—ã
        OnAction += (msg) => Console.WriteLine($"üêï {msg}");
        OnFeed += (food) => Console.WriteLine($"–°–æ–±–∞–∫–∞ {Name} –µ—Å—Ç {food} —Å —Ä–∞–¥–æ—Å—Ç—å—é!");
    }

    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã
    public void Fetch()
    {
        RaiseAction($"{Name} –ø—Ä–∏–Ω–æ—Å–∏—Ç {FavoriteToy}");
    }

    public void Bark()
    {
        RaiseAction($"{Name} –≥—Ä–æ–º–∫–æ –ª–∞–µ—Ç");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–ü–æ—Ä–æ–¥–∞: {Breed}, –î—Ä–µ—Å—Å–∏—Ä–æ–≤–∞–Ω: {IsTrained}, –õ—é–±–∏–º–∞—è –∏–≥—Ä—É—à–∫–∞: {FavoriteToy}");
    }
}

public class Cat : Animal
{
    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã
    public int LivesCount { get; set; }
    public bool LikesCuddles { get; set; }
    public string FurColor { get; set; }

    public Cat(string name, int age, double weight, int livesCount, bool likesCuddles, string furColor) 
        : base(name, age, weight, "–î–æ–º")
    {
        LivesCount = livesCount;
        LikesCuddles = likesCuddles;
        FurColor = furColor;

        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏—è
        OnAction += (msg) => Console.WriteLine($"üêà {msg}");
        OnFeed += (food) => Console.WriteLine($"–ö–æ—à–∫–∞ {Name} –≤–∞–ª—å—è–∂–Ω–æ –µ—Å—Ç {food}");
    }

    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã
    public void Purr()
    {
        RaiseAction($"{Name} –º—É—Ä–ª—ã—á–µ—Ç");
    }

    public void ClimbTree()
    {
        RaiseAction($"{Name} –ª–∞–∑–∞–µ—Ç –ø–æ –¥–µ—Ä–µ–≤—å—è–º");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–ñ–∏–∑–Ω–∏: {LivesCount}, –õ—é–±–∏—Ç –ª–∞—Å–∫—É: {LikesCuddles}, –¶–≤–µ—Ç —à–µ—Ä—Å—Ç–∏: {FurColor}");
    }
}

public class Bird : Animal
{
    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã
    public double Wingspan { get; set; }
    public bool CanFly { get; set; }
    public string FeatherColor { get; set; }

    public Bird(string name, int age, double weight, double wingspan, bool canFly, string featherColor) 
        : base(name, age, weight, "–ù–µ–±–æ")
    {
        Wingspan = wingspan;
        CanFly = canFly;
        FeatherColor = featherColor;

        // –ü–æ–¥–ø–∏—Å–∫–∞ –Ω–∞ —Å–æ–±—ã—Ç–∏—è
        OnAction += (msg) => Console.WriteLine($"üê¶ {msg}");
        OnFeed += (food) => Console.WriteLine($"–ü—Ç–∏—Ü–∞ {Name} –∫–ª—é–µ—Ç {food}");
    }

    // –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã
    public void Fly()
    {
        if (CanFly)
            RaiseAction($"{Name} –ª–µ—Ç–∞–µ—Ç —Å —Ä–∞–∑–º–∞—Ö–æ–º –∫—Ä—ã–ª—å–µ–≤ {Wingspan}—Å–º");
        else
            RaiseAction($"{Name} –Ω–µ —É–º–µ–µ—Ç –ª–µ—Ç–∞—Ç—å");
    }

    public void Sing()
    {
        RaiseAction($"{Name} –∫—Ä–∞—Å–∏–≤–æ –ø–æ–µ—Ç");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"–†–∞–∑–º–∞—Ö –∫—Ä—ã–ª—å–µ–≤: {Wingspan}—Å–º, –£–º–µ–µ—Ç –ª–µ—Ç–∞—Ç—å: {CanFly}, –¶–≤–µ—Ç –ø–µ—Ä—å–µ–≤: {FeatherColor}");
    }
}

class Program
{
    static void Main()
    {
        // –ö–æ–ª–ª–µ–∫—Ü–∏—è –∂–∏–≤–æ—Ç–Ω—ã—Ö
        List<Animal> animals = new List<Animal>();

        // –°–æ–∑–¥–∞–µ–º –∂–∏–≤–æ—Ç–Ω—ã—Ö —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏
        animals.Add(new Dog("–ë–æ–±–∏–∫", 3, 25.5, "–û–≤—á–∞—Ä–∫–∞", true, "–ú—è—á–∏–∫"));
        animals.Add(new Cat("–ú—É—Ä–∫–∞", 2, 4.2, 9, true, "–°–µ—Ä—ã–π"));
        animals.Add(new Bird("–ö–µ—à–∞", 1, 0.3, 15.5, true, "–ó–µ–ª–µ–Ω—ã–π"));
        animals.Add(new Dog("–†–µ–∫—Å", 5, 30.1, "–î–æ–±–µ—Ä–º–∞–Ω", false, "–ö–æ—Å—Ç—å"));

        // –ì–ª–æ–±–∞–ª—å–Ω—ã–π –æ–±—Ä–∞–±–æ—Ç—á–∏–∫ —Å–æ–±—ã—Ç–∏–π
        AnimalAction globalHandler = (msg) => Console.WriteLine($"üì¢ {DateTime.Now:HH:mm:ss}: {msg}");

        // –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º –≤—Å–µ—Ö –∂–∏–≤–æ—Ç–Ω—ã—Ö –Ω–∞ –≥–ª–æ–±–∞–ª—å–Ω—ã–π –æ–±—Ä–∞–±–æ—Ç—á–∏–∫
        foreach (var animal in animals)
        {
            animal.OnAction += globalHandler;
        }

        Console.WriteLine("=== –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –ñ–ò–í–û–¢–ù–´–• ===");
        // –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—è–º–±–¥—É –¥–ª—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏
        animals.ForEach(animal => animal.DisplayInfo());

        Console.WriteLine("\n=== –î–ï–ô–°–¢–í–ò–Ø –ñ–ò–í–û–¢–ù–´–• ===");
        // –í—ã–∑—ã–≤–∞–µ–º —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–µ—Ç–æ–¥—ã —á–µ—Ä–µ–∑ –¥–µ–ª–µ–≥–∞—Ç—ã
        Action<Animal>[] actions = {
            a => a.Move(),
            a => a.Sleep(),
            a => {
                if (a is Dog dog) dog.Fetch();
                else if (a is Cat cat) cat.Purr();
                else if (a is Bird bird) bird.Fly();
            }
        };

        // –ü—Ä–∏–º–µ–Ω—è–µ–º –¥–µ–π—Å—Ç–≤–∏—è –∫ –∫–∞–∂–¥–æ–º—É –∂–∏–≤–æ—Ç–Ω–æ–º—É
        foreach (var animal in animals)
        {
            Console.WriteLine($"\n--- {animal.Name} ---");
            foreach (var action in actions)
            {
                action(animal);
            }
        }

        Console.WriteLine("\n=== –ö–û–†–ú–õ–ï–ù–ò–ï –ñ–ò–í–û–¢–ù–´–• ===");
        // –ö–æ—Ä–º–∏–º –∂–∏–≤–æ—Ç–Ω—ã—Ö
        animals[0].Feed("–º—è—Å–æ");      // –°–æ–±–∞–∫–∞
        animals[1].Feed("—Ä—ã–±–∞");      // –ö–æ—à–∫–∞  
        animals[2].Feed("–∑–µ—Ä–Ω–æ");     // –ü—Ç–∏—Ü–∞
        animals[3].Feed("–∫–æ—Å—Ç–æ—á–∫–∏");  // –°–æ–±–∞–∫–∞

        Console.WriteLine("\n=== –ü–û–ò–°–ö –ü–û –ö–†–ò–¢–ï–†–ò–Ø–ú ===");
        // –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—è–º–±–¥—ã –¥–ª—è –ø–æ–∏—Å–∫–∞
        var heavyAnimals = animals.FindAll(a => a.Weight > 10);
        Console.WriteLine("–¢—è–∂–µ–ª—ã–µ –∂–∏–≤–æ—Ç–Ω—ã–µ (>10–∫–≥):");
        heavyAnimals.ForEach(a => Console.WriteLine($"- {a.Name} ({a.Weight}–∫–≥)"));

        var youngAnimals = animals.FindAll(a => a.Age < 4);
        Console.WriteLine("\n–ú–æ–ª–æ–¥—ã–µ –∂–∏–≤–æ—Ç–Ω—ã–µ (<4 –ª–µ—Ç):");
        youngAnimals.ForEach(a => Console.WriteLine($"- {a.Name} ({a.Age} –ª–µ—Ç)"));
    }
}
```
