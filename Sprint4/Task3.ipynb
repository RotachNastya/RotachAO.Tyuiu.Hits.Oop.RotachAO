```
// Базовый класс для всех животных
public class Animal
{
    // Свойства животного
    public string Name { get; set; }  // Имя
    public int Age { get; set; }      // Возраст
    public string Color { get; set; } // Цвет
    public double Weight { get; set; } // Вес
    public string Diet { get; set; }  // Питание

    // Конструктор класса Animal
    public Animal(string name, int age, string color, double weight, string diet)
    {
        Name = name;
        Age = age;
        Color = color;
        Weight = weight;
        Diet = diet;
    }

    // Виртуальные методы для полиморфизма
    public virtual string Sound() => "Издает звук";       // Общий звук
    public virtual string Eat() => $"{Name} ест.";        // Как животное ест
    public virtual string Sleep() => $"{Name} спит.";     // Сон
    public virtual string Move() => $"{Name} двигается.";// Движение

    // Переопределение ToString для красивого вывода информации
    public override string ToString()
    {
        return $"Имя: {Name}, Возраст: {Age}, Цвет: {Color}, Вес: {Weight} кг, Питание: {Diet}";
    }
}

// Класс Собака — наследуется от Animal
public class Dog : Animal
{
    public string Breed { get; set; }    // Порода
    public bool IsTrained { get; set; }  // Дрессированность

    // Конструктор класса Dog
    public Dog(string name, int age, string color, double weight, string diet, string breed, bool isTrained)
        : base(name, age, color, weight, diet)
    {
        Breed = breed;
        IsTrained = isTrained;
    }

    // Переопределение методов для конкретного животного
    public override string Sound() => "Гав-гав!";
    public override string Move() => $"{Name} бегает.";

    // Дополнительный метод для собаки
    public string Guard() => $"{Name} охраняет дом.";

    // Расширяем ToString дополнительной информацией
    public override string ToString() => base.ToString() + $", Порода: {Breed}, Дрессирован: {IsTrained}";
}

// Абстрактный класс Млекопитающее — наследуется от Animal
public abstract class Mammal : Animal
{
    public bool HasFur { get; set; }        // Есть ли шерсть
    public int GestationPeriod { get; set; } // Период беременности в днях

    // Конструктор класса Mammal
    public Mammal(string name, int age, string color, double weight, string diet, bool hasFur, int gestationPeriod)
        : base(name, age, color, weight, diet)
    {
        HasFur = hasFur;
        GestationPeriod = gestationPeriod;
    }

    // Метод кормления молоком
    public string FeedMilk() => $"{Name} кормит молоком.";
}

// Класс Кошка — наследуется от Mammal
public class Cat : Mammal
{
    public Cat(string name, int age, string color, double weight, string diet, bool hasFur, int gestationPeriod)
        : base(name, age, color, weight, diet, hasFur, gestationPeriod) { }

    public override string Sound() => "Мяу!";
    public override string Move() => $"{Name} крадётся.";
    public string Climb() => $"{Name} залезает на дерево.";

    public override string ToString() =>
        base.ToString() + $", Имеет шерсть: {HasFur}, Беременность: {GestationPeriod} дней";
}

// Интерфейс для домашних животных
public interface IPet
{
    string Play(); // Метод игры
}

// Интерфейс для обучаемых животных
public interface ITrainable
{
    string Train(); // Метод дрессировки
}

// Класс Умная собака — наследуется от Dog и реализует интерфейсы
public class SmartDog : Dog, IPet, ITrainable
{
    public SmartDog(string name, int age, string color, double weight, string diet, string breed, bool isTrained)
        : base(name, age, color, weight, diet, breed, isTrained) { }

    public string Play() => $"{Name} играет с мячиком.";
    public string Train() => $"{Name} проходит дрессировку.";
}

// Generic-класс Вольер для животных
public class Enclosure<T> where T : Animal
{
    private System.Collections.Generic.List<T> animals = new System.Collections.Generic.List<T>();
    public string Type { get; set; } // Тип вольера

    public Enclosure(string type)
    {
        Type = type;
    }

    // Добавление животного в вольер
    public void AddAnimal(T animal)
    {
        animals.Add(animal);
        System.Console.WriteLine($"{animal.Name} помещен в {Type}");
    }

    // Заставить всех животных издать звук
    public void MakeAllSounds()
    {
        System.Console.WriteLine($"Звуки в {Type}:");
        foreach (var a in animals)
        {
            System.Console.WriteLine($"{a.Name}: {a.Sound()}");
        }
    }

    // Переопределение ToString для отображения всех животных в вольере
    public override string ToString()
    {
        string result = $"Тип вольера: {Type}\n";
        foreach (var a in animals)
            result += "\t" + a + "\n";
        return result;
    }
}

// Точка входа программы

    {
        // Создаем объекты животных
        var dog = new Dog("Рекс", 3, "Коричневый", 15.5, "Мясо", "Такса", true);
        var cat = new Cat("Барсик", 2, "Белый", 4.2, "Мясо", true, 63);
        var smartDog = new SmartDog("Макс", 2, "Черный", 12.0, "Мясо", "Лабрадор", true);

        System.Console.WriteLine("=== Демонстрация полиморфизма ===");

        // Список животных разного типа
        var animals = new System.Collections.Generic.List<Animal> { dog, cat, smartDog };

        foreach (var a in animals)
        {
            System.Console.WriteLine($"{a.GetType().Name} {a.Name}:");
            System.Console.WriteLine($"\tЗвук: {a.Sound()}");   // Полиморфизм — вызывается метод конкретного типа
            System.Console.WriteLine($"\tЕст: {a.Eat()}");
            System.Console.WriteLine($"\tСпит: {a.Sleep()}");
            System.Console.WriteLine($"\tДвижется: {a.Move()}");
            System.Console.WriteLine($"\tИнфо: {a}");

            // Специфичные методы только для собак
            if (a is Dog d)
                System.Console.WriteLine($"\tОхрана: {d.Guard()}");
            System.Console.WriteLine();
        }

        // Создаем вольер для животных
        var predatorEnclosure = new Enclosure<Animal>("Вольер для хищников");
        predatorEnclosure.AddAnimal(dog);
        predatorEnclosure.AddAnimal(cat);
        predatorEnclosure.MakeAllSounds();
        System.Console.WriteLine(predatorEnclosure);
    }


```
